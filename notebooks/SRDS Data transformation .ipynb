```python
# import packages
import pandas as pd
import numpy as np
```

```python
# importing customer identity data 
customer_info = pd.read_excel('srds_customer_identity_adjustment_20220509.xlsx', 'created_proxy_customer_id')
customer_info.head()
```

```python
# importing srds data
srds_data = pd.read_excel('S.R.D.S+Ltd_Sales+by+Customer+Summary (17).xlsx')
srds_data.head()
```

```python
# remove first 3 rows
srds_data = srds_data.iloc[3:].reset_index(drop = True)

# add value to first row, first column
srds_data.iloc[0,0] = 'vendors'

# identify index of row TOTAL
condition_index = srds_data[srds_data['S.R.D.S Ltd'] == 'TOTAL'].index[0]

# slice DataFrame to keep only rows before condition index
srds_data = srds_data.iloc[:condition_index, :]

# remove rows where there is a null value in the first column
#srds_data = srds_data.dropna(axis=0, subset=['S.R.D.S Ltd'])

# setting the first row as the header of the dataframe
srds_data.columns =  srds_data.iloc[0]

# dropping the first row and re-indexing the data
srds_data.drop(0, axis = 'rows', inplace = True)
srds_data.reset_index(drop = True, inplace = True)

srds_data.head()
```

```python
#create dataframe for the vendors only
vendors_column = pd.DataFrame(srds_data['vendors'])

# remove numbers and slash at the end of the name
vendors_column['names'] = vendors_column['vendors'].str.replace(r'\s*\/\s*\d+$', '', regex=True)
vendors_column['names'] = vendors_column['names'].str.replace(r'\s*\d+$', '', regex=True)
vendors_column['names_no_slash'] = vendors_column['names'].str.replace(r'\s*\/', '', regex=True)
vendors_column
```

```python
pd.set_option('display.float_format', lambda x: '%.0f' % x)
```

```python
# merging with customer info data to get the identifiers
merged_df = pd.merge(vendors_column, customer_info, left_on='names_no_slash', right_on='vendors', how = 'left')
merged_df
```

```python
#getting only names and customer id
first_part = merged_df[['names', 'proxy_customer_id']].copy()
# rename columns
first_part.rename(columns = {'names':'vendors', 'proxy_customer_id':'customer_id'}, inplace = True)

first_part
```

```python
# getting the actual data per month
second_part = srds_data.iloc[: , 1:]
```

```python
final_data = pd.concat([first_part, second_part], axis=1)
final_data
```

```python
# saving the excel
final_data.to_excel('srds_feb_2023.xlsx', index=False)
```
